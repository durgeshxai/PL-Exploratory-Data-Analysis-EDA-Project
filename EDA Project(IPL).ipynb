### 7.IPL Finals venues and winners along with the number of wins.


```python
finals = data_1[data_1['date'] == data_1.groupby('season')['date'].transform('max')]
finals_winners = finals.groupby(['venue', 'winner']).size().reset_index(name='wins')

print("IPL final venues and winner and no. of Win:")
finals_winners
```

### 8.The number of seasons won by each team


```python
season_win_by_team= data_1.groupby('season')['winner'].last()
winners = season_win_by_team.value_counts()

plt.figure(figsize=(4, 5))
winners.plot(kind='line', color='blue')
sns.barplot(x=winners.index,y=winners)
plt.title('The number of seasons won by each team\n',fontdict={"fontsize":18,"fontweight":"bold"})
plt.xlabel("\nWinners",fontdict={"fontsize":15,"fontweight":"bold"})
plt.ylabel("NO. of wins\n",fontdict={"fontsize":15,"fontweight":"bold"})
plt.xticks(rotation=60)
plt.show()
```

### 11.Decision in every toss either fielding or batting

```python
decision_for_toss = data_1['toss_decision'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=decision_for_toss.index, y=decision_for_toss.values, palette='Set1')
plt.title('Toss Decision in IPL Matches\n',fontdict={"fontsize":15,"fontweight":"bold"})
plt.xlabel('\nToss Decision',fontdict={"fontsize":15,"fontweight":"bold"})
plt.ylabel('Count\n',fontdict={"fontsize":15,"fontweight":"bold"})
```

### 12.What is the probability of winning a match if the toss was won?

```python
data_1['toss_match_winner'] = data_1['toss_winner'] == data_1['winner']
plt.figure(figsize=(8, 6))
sns.countplot(x='toss_match_winner', data=data_1)
plt.title('Probability of winning a match\n',fontdict={"fontsize":18,"fontweight":"bold"})
plt.xlabel('\nToss Winner',fontdict={"fontsize":18,"fontweight":"bold"})
plt.ylabel('Count\n',fontdict={"fontsize":18,"fontweight":"bold"})
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'],fontdict={"fontsize":14,"fontweight":"bold"})
plt.show()
```

### 13.The total runs by fours hit and the total number of fours hit by each team

```python
merge_data = pd.merge(data_1,data_2, left_on='id', right_on='match_id', how='inner')
total_runs = merge_data[merge_data['batsman_runs'] == 4]['batsman_runs'].sum()
fours_hit = merge_data[merge_data['batsman_runs'] == 4].groupby('batting_team').size()

plt.figure(figsize=(10, 6))
fours_hit.plot(kind='bar', color='red')
plt.title('Total no. of 4 hit by each team\n',fontdict={"fontsize":15,"fontweight":"bold"})
plt.xlabel('Teams',fontdict={"fontsize":15,"fontweight":"bold"})
plt.ylabel('Total fours hit\n',fontdict={"fontsize":15,"fontweight":"bold"})
plt.xticks(rotation=60)
plt.grid(axis='y', linestyle='--', alpha=0.8)
for i, value in enumerate(fours_hit):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

### 14.The fours hit by players

```python
merge_data = pd.merge(data_1,data_2, left_on='id', right_on='match_id', how='inner')
hit_player_4 = merge_data[merge_data['batsman_runs'] == 4].groupby('batsman').size()
plt.figure(figsize=(12, 6))
hit_player_4.sort_values(ascending=False)[:20].plot(kind='bar', color='blue')
plt.title('Total Fours Hit by Players\n',fontdict={"fontsize":20,"fontweight":"bold"})
plt.xlabel('Player',fontdict={"fontsize":15,"fontweight":"bold"})
plt.ylabel('Total Fours Hit\n',fontdict={"fontsize":15,"fontweight":"bold"})
plt.xticks(rotation=60)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, value in enumerate(hit_player_4.sort_values(ascending=False)[:20]):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### 15.The number of fours hit in each season

```python
merge_data = pd.merge(data_1,data_2, left_on='id', right_on='match_id', how='inner')
fours_hit_per_season = merge_data[merge_data['batsman_runs'] == 4].groupby('season').size()
plt.figure(figsize=(10, 6))
fours_hit_per_season.plot(kind='bar', color='blue')
plt.title('Number of Fours Hit in Each Season\n',fontdict={"fontsize":20,"fontweight":"bold"})
plt.xlabel('\n Seasons',fontdict={"fontsize":15,"fontweight":"bold"})
plt.ylabel('Number of Fours Hit\n',fontdict={"fontsize":15,"fontweight":"bold"})
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, value in enumerate(fours_hit_per_season):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### 16.The total runs by the sixes hit and the number of sixes hit by each team

```python
merge_data = pd.merge(data_1,data_2, left_on='id', right_on='match_id', how='inner')
total_runs = merge_data[merge_data['batsman_runs'] == 6]['batsman_runs'].sum()
sixes_hit = merge_data[merge_data['batsman_runs'] == 6].groupby('batting_team').size()

plt.figure(figsize=(10, 6))
fours_hit.plot(kind='bar', color='orange')
plt.title('Total no. of 6 hit by each team\n',fontdict={"fontsize":20,"fontweight":"bold"})
plt.xlabel('Team',fontdict={"fontsize":15,"fontweight":"bold"})
plt.ylabel('Total sixes hit\n',fontdict={"fontsize":15,"fontweight":"bold"})
plt.xticks(rotation=60)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, value in enumerate(fours_hit):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### 17.The number of sixes hit in each season

```python
merge_data = pd.merge(data_1,data_2, left_on='id', right_on='match_id', how='inner')
sixes_hit_per_season = merge_data[merge_data['batsman_runs'] == 6].groupby('season').size()
plt.figure(figsize=(10, 6))
fours_hit_per_season.plot(kind='bar', color='royalblue')
plt.title('Number of sixes Hit in Each Season\n',fontdict={"fontsize":20,"fontweight":"bold"})
plt.xlabel('\n Seasons',fontdict={"fontsize":15,"fontweight":"bold"})
plt.ylabel('Number of sixes Hit\n',fontdict={"fontsize":15,"fontweight":"bold"})
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, value in enumerate(fours_hit_per_season):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

### 18.The top 10 leading run scorers in IPL

```python
merge_data = pd.merge(data_1,data_2, left_on='id', right_on='match_id', how='inner')
total_runs_by_batsman=merge_data.groupby('batsman')['batsman_runs'].sum()
top_10_run_scorers = total_runs_by_batsman.sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
top_10_run_scorers.plot(kind='bar', color='blue')
plt.title('Top 10 Leading Run Scorers in IPL\n',fontdict={"fontsize":20,"fontweight":"bold"})
plt.xlabel('Players',fontdict={"fontsize":16,"fontweight":"bold"})
plt.ylabel('Total Runs\n',fontdict={"fontsize":16,"fontweight":"bold"})
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, value in enumerate(top_10_run_scorers):
    plt.text(i, value + 5, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
